# Scrapping car details from NEW DELHI

```python
driver=webdriver.Chrome(r"C:\Users\Ritu Dalal\Chromedriver\chromedriver.exe") 


driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://www.cars24.com/buy-used-car?sort=P&storeCityId=2&pinId=110001"
driver.get(url)
```

```python
car_name=[]
car_model=[]
car_price=[]
car_info=[]
zero_payment=[]
```

```python
names=driver.find_elements_by_xpath("//h2[@class='_3FpCg']")
for i in names:
    if i.text is None :
        car_name.append("--") 
    else:
        car_name.append(i.text)

        
model=driver.find_elements_by_xpath("//p[@class='cvakB']")
for i in model:
    if i.text is None :
        car_model.append("--") 
    else:
        car_model.append(i.text)     
        
price=driver.find_elements_by_xpath("//div[@class='_7udZZ']")
for i in price:
    if i.text is None :
        car_price.append("--") 
    else:
        car_price.append(i.text)

        
info=driver.find_elements_by_xpath("//ul[@class='bVR0c']") 
for i in info:
    if i.text is None :
        car_info.append("--") 
    else:
        car_info.append(i.text)

payment=driver.find_elements_by_xpath("//div[@class='_2HFRN']")  
for i in payment:
    if i.text is None :
        zero_payment.append("--") 
    else:
        zero_payment.append(i.text)
```

```python
#creating a dataframe
df_delhi=pd.DataFrame({'car_name':car_name[:1000],
                'car_model':car_model[:1000],
                'car_price':car_price[:1000],
                'zero_payment':zero_payment[:1000],       
                'car_info':car_info[:1000]})
                
#printing dataframe
df_delhi
```

```python
df_delhi.to_excel('car_data.xls', index=False)
```
